```python
#Deplasmanda oynanan total maçları bar grafiği ile inceleyelim
data['Away'].value_counts().plot.bar(figsize=(10,5))
```

```python
#Deplasman gollerinin ortalamasını aldık
away_mean=['AwayGoals']
data[away_mean].mean()
```

```python
total_home_goals=['HomeGoals']
data[total_home_goals].sum()
```

```python
data['Home'].value_counts().plot.bar(figsize=(10,5))
```

```python
data.groupby(['Home'])['HomeGoals'].mean().sort_values(ascending=False).plot.bar(figsize=(10,5),color='Red')
```

```python
data.groupby(['Home'])['HomeGoals'].sum().sort_values(ascending=False).plot.bar(figsize=(10,5),color='Red')
```

```python
data.groupby(['Away'])['AwayGoals'].mean().sort_values(ascending=False).plot.bar(figsize=(10,5),color='Red')
```

```python
#Deplasmanda atılan gol toplamını azalan sıraya göre grafiğe döktük burada takımların attıkları toplam golleri görüyoruz(deplasman)
data.groupby(['Away'])['AwayGoals'].sum().sort_values(ascending=False).plot.bar(figsize=(10,5),color='Red')
```
